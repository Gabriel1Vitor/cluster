```python
bag_tranf=CountVectorizer(min_df=4,stop_words="english",ngram_range=(3,3))
mx=bag_tranf.fit_transform(text["txt"])
mx.shape
```

```python
for k in range(2,10):
  cluster = make_pipeline(MaxAbsScaler(),KMeans(n_clusters=k,random_state=0))
  cluster.fit(mx)
  p = cluster.predict(mx)
  sil = silhouette_score(mx,p)
  print("K = {} = Silhouette = {}" . format(k , sil))
#silhouette_score o mais usado para cluster
```

```python
terms=bag_tranf.get_feature_names()
```

```python
k = 8
cluster=make_pipeline(MaxAbsScaler(),KMeans(n_clusters=k,random_state=0))
cluster.fit(mx)
p=cluster.predict(mx)

for c in np.unique(p):
  print("\nCluster {} - Size {} ".format(c,(p == c).sum()))
  rank=pd.Series(np.array(mx[p==c].mean(axis=0)).squeeze(),index=terms).sort_values().tail(20)
  print(rank)
```

```python
centroid=cluster.named_steps["kmeans"].cluster_centers_
for c in range(centroid.shape[0]):
  print("\nCluster {}".format(c))
  rank = pd.Series(centroid[c,:],index=terms).sort_values().tail(20)
  print(rank)
```

```python
text["cluster"]=p
```
